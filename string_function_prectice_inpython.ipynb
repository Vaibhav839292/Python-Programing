String Function

```python
#endswith() - Returns true if the string ends with the specified value

a ="Haryy Potter"
#print(a.endswith("r"))
print(a.endswith("t",6,9))

```

```python
#startwith() - returns true if the string starts with the specified value

a = "Harry potter"
print (a.startswith("H"))
```

```python
#swapcase() - swap cases , lower case become upper case and vice versa

a ="Harry Potter"
print(a.swapcase())
```

```python
#strip()- Retruns a trimmed version of the string

a = "Harry  potter     "

print(a.strip())
```

```python
#split()- splits the string at the specified separator, and returns a list

a ="#OOFD#BRB#OMW#TB"
print(a.split("#"))
```

```python
#ljust - returns a left justified version of the string

a = "Harry Potter"

x = a.ljust(20)

print (x,"is my favroit movie")
```

```python
#rjust - return a right

a = "hrry potter"
x = a.rjust(20," ")

print(x,"is my fav movie")
```

```python
#replace() - returns a string where a specified value is replaced with a specified value
a ="my name is jhon"
print(a)
print(a.replace("jhone", "vaibhav"))
```

```python
#rindex()

a ="harry potter and the prisoner of azkaban"

print(a.rindex("prisoner"))
```

```python
#rfind

a ="harry potter and the prisoner of azkaban"
print(a.rfind("prisoner"))
```
